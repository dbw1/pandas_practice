```python
def custom_heuristic(file_path):

    predictions = {}
    df = pd.read_csv(file_path)
    for passenger_index, passenger in df.iterrows():
        passenger_id = passenger['PassengerId']
        if (passenger['Sex'] == 'female' and passenger['Pclass'] in [1,2])or (passenger['Age']<6) :
            predictions[passenger_id] = 1
        else:
            predictions[passenger_id] = 0
    return predictions
```

```python
pred = custom_heuristic(file_path)
```

```python
pred_df = pd.DataFrame(list(pred.items()), columns=['PassengerId','Survived'])
```

```python
from __future__ import division
def test_accuracy(pred_dict, test_data_df):
    pred_df = pd.DataFrame(list(pred.items()), columns=['PassengerId','Survived'])
    total_correct_count = 0
    total_rows = test_data_df.shape[0]
    for passenger_index, passenger in pred_df.iterrows():
        if passenger['Survived'] == test_data_df['Survived'][passenger_index]:
            total_correct_count += 1
    
    return total_correct_count / total_rows
```
